```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if(counter != 3):
            L.append(i['name'])
        else:
            break
    counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def convert3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies
```

```python
def SpaceRemover(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ", ""))
    return L
```

```python
movies.info()
```

```python
movies['genres'] = movies['genres'].apply(SpaceRemover)
movies['cast'] = movies['cast'].apply(SpaceRemover)
movies['crew'] = movies['crew'].apply(SpaceRemover)
movies['keywords'] = movies['keywords'].apply(SpaceRemover)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new_df = movies[['movie_id', 'title', 'tag']]
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectorizer = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectorizer.shape
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(enumerate(similarity[2]))
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
